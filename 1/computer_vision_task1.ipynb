```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib as mpl
```

```python
img_original = cv2.imread('/content/water-2208842_1280.jpg')
img_original = cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB)
plt.imshow(img_original)
plt.show()
```

```python
print("высота:", img_original.shape[0], "ширина:", img_original.shape[1])
print("среднее значение интенсивности пикселя:", np.mean(img_original))
```

```python
def exposure(image, p):
    img_result = np.round(image.copy() * (2 ** p),0)
    img_result = img_result.astype(int)
    for i in range(img_result.shape[0]):
      for j in range(img_result.shape[1]):
        for k in range(img_result.shape[2]):
          if img_result[i][j][k] > 255:
            img_result[i][j][k] = 255
    return img_result


# менять значение p по заданию
img_result = exposure(img_original, -0.22)

print("Среднее значение интенсивности пикселя исходного изображения:", np.mean(img_original))
print("Среднее значение интенсивности пикселя модифицированного изображения:", np.mean(img_result))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 10))
axes[0].imshow(img_original)
axes[1].imshow(img_result)
```

```python


# Коэффициент для изменения экспозиции (1.5 увеличит экспозицию в 1.5 раза)
gamma = 2**0.46

# Преобразование экспозиции
adjusted_image = cv2.convertScaleAbs(img_original, alpha=gamma, beta=0)

# Отображение изображения до и после изменения экспозиции
print("Среднее значение интенсивности пикселя исходного изображения:", np.mean(img_original))
print("Среднее значение интенсивности пикселя модифицированного изображения:", np.mean(adjusted_image))

fig, axes = plt.subplots(1, 2, figsize=(20, 10))
axes[0].imshow(img_original)
axes[1].imshow(adjusted_image)


```

```python
# Коэффициент для изменения экспозиции (1.5 увеличит экспозицию в 1.5 раза)
gamma = 2**-0.22

# Преобразование экспозиции
adjusted_image = cv2.convertScaleAbs(img_original, alpha=1, beta=gamma)

# Отображение изображения до и после изменения экспозиции
print("Среднее значение интенсивности пикселя исходного изображения:", np.mean(img_original))
print("Среднее значение интенсивности пикселя модифицированного изображения:", np.mean(adjusted_image))

fig, axes = plt.subplots(1, 2, figsize=(20, 10))
axes[0].imshow(img_original)
axes[1].imshow(adjusted_image)
```
